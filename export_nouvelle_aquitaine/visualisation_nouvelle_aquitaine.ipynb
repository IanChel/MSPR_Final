# Visualisation des résultats d'élections - Nouvelle-Aquitaine (2012 - 2017)
Ce notebook permet d'explorer et de visualiser les données électorales extraites pour les deux tours.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuration visuelle
sns.set_theme(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)
```

## 1. Chargement des données

```python
df_tour1 = pd.read_csv('nouvelle_aquitaine_2012_2017_tour1.csv')
df_tour2 = pd.read_csv('nouvelle_aquitaine_2012_2017_tour2.csv')

print(f"Tour 1 : {df_tour1.shape[0]} lignes")
print(f"Tour 2 : {df_tour2.shape[0]} lignes")

display(df_tour1.head(3))
```

## 2. Visualisation : Abstention par année

```python
# Comparaison de l'abstention (% Abs/Ins) au 1er et 2ème tour
fig, axes = plt.subplots(1, 2, figsize=(14, 5), sharey=True)

sns.boxplot(data=df_tour1, x='Année', y='% Abs/Ins', ax=axes[0], palette="Blues")
axes[0].set_title('Distribution de l\'abstention - Tour 1')
axes[0].set_ylabel('% Abstention / Inscrits')

sns.boxplot(data=df_tour2, x='Année', y='% Abs/Ins', ax=axes[1], palette="Oranges")
axes[1].set_title('Distribution de l\'abstention - Tour 2')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

## 3. Visualisation : Inscrits vs Votants par Département (2017)

```python
# Focus sur 2017, Tour 1
df_2017_t1 = df_tour1[df_tour1['Année'] == 2017]

# Agréger par département
dept_stats = df_2017_t1.groupby('Libellé du département')[['Inscrits', 'Votants']].sum().reset_index()

dept_stats.set_index('Libellé du département').plot(kind='bar', stacked=False, figsize=(14, 6))
plt.title('Comparaison Inscrits vs Votants par département (Tour 1 - 2017)')
plt.ylabel('Nombre de personnes')
plt.xlabel('Département')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
